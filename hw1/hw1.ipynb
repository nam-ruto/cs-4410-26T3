```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## Task (b): Load the dataset into a pandas DataFrame

```python
# Load the dataset using the first column as the row index
df = pd.read_csv('diamonds.csv', index_col=0)
print(f"Dataset loaded successfully! Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

## Task (e): Descriptive statistics for numerical columns

```python
# Select numerical columns: carat, depth, table, price, x, y, z
numerical_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
df[numerical_cols].describe()
```

## Task (f): Descriptive statistics for categorical columns

```python
# Descriptive statistics for categorical columns: cut, color, clarity
categorical_cols = ['cut', 'color', 'clarity']

for col in categorical_cols:
    print(f"\n{col.upper()}:")
    print(df[col].describe())
    print("-" * 50)
```

## Task (g): Unique category values

```python
# Find unique values for each categorical column
for col in categorical_cols:
    unique_vals = df[col].unique()
    print(f"\n{col.upper()} unique values ({len(unique_vals)}):")
    print(sorted(unique_vals))
    print("-" * 50)
```

## Task (h): Histograms for numerical columns

```python
# Enable matplotlib support in IPython
%matplotlib inline

# Create histograms for all numerical columns
df[numerical_cols].hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```
